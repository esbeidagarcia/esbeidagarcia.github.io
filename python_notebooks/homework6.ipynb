```python
import altair as alt
import pandas as pd
```

```python
url  = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
```

```python
data = pd.read_csv(url)
```

```python
building_type_count = data['Agency Name'].value_counts().reset_index()
building_type_count.columns = ['Building Type', 'Count']

# Create Altair bar chart
chart1 = alt.Chart(building_type_count).mark_bar().encode(
    x=alt.X('Count:Q', title='Number of Buildings'),
    y=alt.Y('Building Type:N', sort='-x', title='Building Type'),
    color=alt.Color('Building Type:N', legend=None),  # Optional: remove legend to avoid redundancy
    tooltip=['Building Type', 'Count']
).properties(
    width=600,
    height=400,
    title="Distribution of Building Types"
)
chart1.show()
```

```python
data['Year Constructed'] = pd.to_numeric(data['Year Constructed'], errors='coerce')  
filtered_data = data.dropna(subset=['Year Constructed', 'Agency Name'])  
filtered_data = filtered_data[filtered_data['Year Constructed'] > 0]  

yearly_construction = (
    filtered_data.groupby(['Year Constructed', 'Agency Name']).size()
    .reset_index(name='BuildingCount')
)

multi_select = alt.selection_multi(fields=['Agency Name'], bind='legend', name="Select Agency")

line_chart = alt.Chart(yearly_construction).mark_line(point=True).encode(
    x=alt.X('Year Constructed:Q', title='Year Constructed', axis=alt.Axis(format='d')),
    y=alt.Y('BuildingCount:Q', title='Number of Buildings'),
    color=alt.Color('Agency Name:N', title='Agency', legend=alt.Legend(title="Agency Name")),
    tooltip=['Year Constructed:Q', 'BuildingCount:Q', 'Agency Name:N'],
    opacity=alt.condition(multi_select, alt.value(1), alt.value(0.2))  
).add_selection(
    multi_select
).properties(
    width=800,
    height=500,
    title="Buildings Constructed Over Time by Agency"
).interactive() 

line_chart.show()
```
